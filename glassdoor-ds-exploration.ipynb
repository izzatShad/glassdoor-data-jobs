```python
#From the results it seems some information in the parenthesis includes infomration that should be in the job description.
#Take note that there are parentheses with the word Sr or possible Jr in it, which should be an exception.

#replace (Sr.) with Sr.
ds_jobs['Job Title'] = ds_jobs.loc[:,'Job Title'].str.replace("(Sr.)", "Sr.")

#replace (Jr.) with Jr.
ds_jobs['Job Title'] = ds_jobs.loc[:,'Job Title'].str.replace("(Jr.)", "Jr.")
```

```python
#I want to investigate what roles have their simplified version as na
ds_jobs.loc[ds_jobs['job_simp']=='na', ['Job Title', 'job_simp']]
```

```python
#The values seem varied, between what could be categorized as a scientist, computer science, software engineer, etc.
#I decided one way to represent the data quickly is by combining the first and last word in the Job title.
na_jobsimp = ds_jobs.loc[ds_jobs['job_simp']=='na', ['Job Title', 'job_simp']]
last_job_title = na_jobsimp['Job Title'].str.split()
last_job_title = last_job_title.str[0] + ' ' + last_job_title.str[-1]
print(last_job_title)
```

```python
#after confirming the contents, replace job_simp where the values are na, as the values from confirmed last_job_title
ds_jobs.loc[ds_jobs['job_simp'] == 'na', 'job_simp'] = ds_jobs['Job Title'].str.split().apply(lambda x: x[0] + ' ' + x[-1] if isinstance(x, list) else x)
```

```python
#printing the updated list of job_simp to check if 'na' is still there
ds_jobs.job_simp.unique()
```

```python
#for consistency purposes let's change everything to lowercase
ds_jobs['job_simp'] = ds_jobs['job_simp'].str.lower()

print(ds_jobs.job_simp.unique())
```

```python
#a lot of the generated entries make sense, like 'medical scientist' or 'statistical scientist' or 'weapons engineer/scientist'
#but some doesn't make much sense, like 'vp, science' or 'research preferences'. These are most likely outliers and would need to be replaced individually

ds_jobs.loc[ds_jobs['job_simp']=='research preferences', ['Job Title', 'job_simp']]
```

```python
#a lot of the generated entries make sense, like 'medical scientist' or 'statistical scientist' or 'weapons engineer/scientist'
#but some doesn't make much sense, like 'vp, science' or 'research preferences'. These are most likely outliers and would need to be replaced individually

# a lot of the entries in job_simp contain the word 'scientist', maybe they're related in a way where it makes sense to categorize all of them as simply 'scientist'
ds_jobs.loc[ds_jobs['job_simp'].str.contains('scientist') & ~ds_jobs['job_simp'].str.contains('data scientist'), ['Job Title', 'job_simp']]
```

```python
#based on the result above, I believe we can re-categorize the different job_simp entries as either 'scientist' or 'data scientist'

#start with replacing items that could be re-categorized as 'scientist'
ds_jobs['job_simp'].replace(['medical scientist', 'human scientist', 'scientist biology', 'purification scientist', 'ngs scientist', 'research scientist', 'scientist cytometry', 'elisa scientist', 'weapons engineer/scientist', 'scientist/research engineering','hydrogen/tritium scientist'], 'scientist',inplace = True)
```

```python
#confirm the change in contents
print(ds_jobs.job_simp.unique())
```

```python
#replace items that can be re-categorized as 'data scientist'
ds_jobs['job_simp'].replace(['computational scientist', 'statistical scientist', 'decision scientist', 'computer scientist', 'applied scientist'], 'data scientist',inplace = True)
```

```python
#confirm the change in contents
print(ds_jobs.job_simp.unique())
```

```python
#based on the result above the entry 'Research Scientist' may be more appropriate
ds_jobs['job_simp'] = ds_jobs.loc[:,'job_simp'].str.replace('research preferences', 'research scientist')
ds_jobs.loc[ds_jobs['job_simp']=='research preferences', ['Job Title', 'job_simp']]

#confirmed that no results left for 'research preferences
```

```python
#define the outliers
job_simp_outliers = ['research research', 'production analysis', 'data instructor', 'engineer or', 'real evidence', 'computer 1', 'environmental science', 'development voltaren', 'principal engineer', 'vp, science', 'applied engineer', 'it science', 'lead shift', 'senior integrity', 'computer processing', 'computer technology', 'information specialist', 'vice management', 'sr.genomics', 'developer science']

#filter the dataframe with the outliers
outlier_job = ds_jobs[ds_jobs['job_simp'].isin(job_simp_outliers)]

#view the data of the filtered rows
outlier_df = outlier_job[['Job Title', 'job_simp']]

print(outlier_df)
```

```python
#replace outlier values
ds_jobs['job_simp'].replace(['research research', 'environmental science', 'development voltaren', 'lead shift', 'senior integrity', 'vice management', 'sr. genomics'], 'scientist',inplace = True)
ds_jobs['job_simp'].replace(['engineer or', 'principal engineer', 'information specialist'], 'data engineer',inplace = True)
ds_jobs['job_simp'].replace(['computer 1', 'vp, science', 'applied engineer', 'computer processing', 'computer technology', 'developer science', 'it science', 'real evidence'], 'data scientist',inplace = True)
ds_jobs['job_simp'].replace(['production analysis'], 'analyst',inplace = True)
```

```python
#review changes
print(ds_jobs.job_simp.unique())

#the results seem to be clean enough in my opinion and all the roles in the list make sense.
```
